```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
Z = linkage(X, 'ward')
iris
```

```python
# Plot the dendogram
plt.title("Iris Dendrogram")
dendrogram(Z)
plt.show()
```

```python
model=AgglomerativeClustering(n_clusters=3)
model.fit(X)
```

```python
labels=model.labels_
labels
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.show()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3)
y_pred=km.fit_predict(X)
y_pred
```

```python
x_K=km.cluster_centers_
x_K
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target #add the target to the DataFrame

iris_df.head(5)
```

```python
from sklearn.cluster import KMeans
data_x = iris_df[["sepal length (cm)", "sepal width (cm)", "petal length (cm)", "petal width (cm)"]] # Select columns using a list of column names
km = KMeans(n_clusters=3)
y_pred = km.fit_predict(data_x) # Use the selected data for clustering
```

```python
iris_df['cluster']=y_pred
```

```python
d1=iris_df[iris_df.cluster==0]
d2=iris_df[iris_df.cluster==1]
d3=iris_df[iris_df.cluster==2]
plt.scatter(d1["sepal length (cm)"],d1["sepal width (cm)"],color="red") # Changed column names to match the DataFrame
plt.scatter(d2["sepal length (cm)"],d2["sepal width (cm)"],color="green") # Changed column names to match the DataFrame
plt.scatter(d3["sepal length (cm)"],d3["sepal width (cm)"],color="blue") # Changed column names to match the DataFrame
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',label='centroid')
plt.plot()
```
